```python
%%writefile my_math.py

# 1. 함수
def multiply(a, b):
    return a * b

# 2. 클래스
class Calculator:
    def add(self, a, b):
        return a + b
```

함수와 클래스 ➞ 모듈 ➞ 패키지 ➞ 라이브러리


```python
# main.py

import my_math # my_math 모듈을 가져옴

# 모듈 안의 함수 사용
result_mul = my_math.multiply(10, 5)
print(f"곱하기 결과: {result_mul}")

# 모듈 안의 클래스 사용
calc = my_math.Calculator()
result_add = calc.add(10, 5)
print(f"더하기 결과: {result_add}")
```

2. 모듈 (Module: 블록들을 모아둔 하나의 상자)
모듈은 파이썬 함수나 클래스, 변수들을 모아 놓은 **하나의 파이썬 파일(.py)**입니다. 코드가 길어지면 관리가 어려워지므로, 기능별로 파일을 나누어 저장하는데 이것이 바로 모듈입니다.


3. 패키지 (Package: 여러 상자를 분류해 둔 서랍장)
패키지는 여러 개의 모듈을 폴더(디렉터리) 구조로 묶어 놓은 것입니다. 프로젝트가 커져서 모듈(.py 파일)이 수십 개가 되면, 이를 성격에 맞게 폴더로 나누어 관리해야 하는데 이것이 패키지입니다.

```python
# string_tools.py

# 1. 모듈 수준의 변수
VERSION = "1.0"
AUTHOR = "Gemini"

# 2. 함수
def reverse_string(text):
    """문자열을 뒤집어서 반환하는 함수입니다."""
    return text[::-1]

def count_vowels(text):
    """문자열에 포함된 모음(a, e, i, o, u)의 개수를 셉니다."""
    vowels = "aeiouAEIOU"
    count = sum(1 for char in text if char in vowels)
    return count

# 3. 클래스
class TextFormatter:
    """텍스트의 포맷을 변경하는 클래스입니다."""
    def __init__(self, text):
        self.text = text

    def to_uppercase(self):
        return self.text.upper()

    def add_exclamation(self):
        return f"{self.text}!!!"
```

4. 라이브러리 (Library: 서랍장들이 모여 있는 거대한 창고)
라이브러리는 모듈과 패키지의 모음으로, 개발자들이 유용하게 사용할 수 있도록 미리 만들어 둔 거대한 코드 덩어리입니다. (종종 패키지와 혼용되어 쓰이기도 하지만, 보통 더 큰 규모를 의미합니다.)

표준 라이브러리: 파이썬을 설치할 때 기본적으로 제공되는 패키지들 (예: math, random, datetime)

외부 라이브러리: 전 세계 개발자들이 만들어 인터넷에 공유한 패키지들. pip install 명령어로 설치하여 사용합니다. (예: pandas, requests, numpy)

```python
# main.py

# 모듈에서 특정 함수와 클래스만 명시적으로 가져옵니다.
from string_tools import count_vowels, TextFormatter

# 'string_tools.' 을 붙일 필요 없이 바로 사용 가능합니다.
my_text = "Beautiful Day"

# 함수 바로 사용
vowel_count = count_vowels(my_text)
print(f"모음의 개수: {vowel_count}") # 출력: 6

# 클래스 바로 사용
formatter = TextFormatter("Awesome")
print(formatter.add_exclamation()) # 출력: Awesome!!!
```

함수/클래스: 코드를 재사용하기 위한 최소 단위

모듈(.py): 함수와 클래스들을 모아둔 단일 파일

패키지(폴더): 여러 모듈을 주제별로 묶어둔 폴더

라이브러리: 패키지와 모듈들의 거대한 집합체 (가져다 쓰기만 하면 되는 도구함)


```python
#모듈은 한마디로 **"관련된 함수, 클래스, 변수들을 모아놓은 하나의 파이썬 파일(.py)"**입니다. 코드를 부품처럼 나눠서 관리하고 싶을 때 사용하죠.

#모듈 만들기 (tools.py)
#사용할 기능들을 파일 하나에 작성합니다.

# tools.py

PI = 3.14

def add(a, b):
    return a + b

class Greet:
    def say_hi(self):
        print("안녕하세요!")
```

```python
#모듈 사용하기 (main.py)
#import 키워드를 사용해 다른 파일의 기능을 가져옵니다.

# main.py

import tools

print(tools.add(5, 3))  # 8 (함수 호출)
print(tools.PI)         # 3.14 (변수 사용)

me = tools.Greet()      # 클래스 사용
me.say_hi()             # "안녕하세요!"
```
