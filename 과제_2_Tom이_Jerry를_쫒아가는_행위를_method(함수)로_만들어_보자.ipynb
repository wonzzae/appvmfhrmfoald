```python
class User:
    def __init__(self, name):

        self.name = name
        self.followers = 0
        self.following = 0

    def follow(self, target_user):

        self.following += 1
        target_user.followers += 1

        print(f"🏃 {self.name}이(가) {target_user.name}을(를) 쫓아갑니다! "
              f"(현재 {self.name}의 추격 횟수: {self.following})")

tom = User("Tom")
jerry = User("Jerry")

print("🏁 추격전 시작!")
for i in range(10):
    tom.follow(jerry)

print("\n--- 최종 결과 ---")
print(f"{tom.name}: 팔로잉(추격 중) {tom.following}명")
print(f"{jerry.name}: 팔로워(도망 중) {jerry.followers}명")
```
